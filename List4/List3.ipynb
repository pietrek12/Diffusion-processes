```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import random
from fitter import Fitter
```

### Random graph

```python
def random_graph(N, p):
    G = nx.Graph()
    for i in range(N):
        G.add_node(i)
    for node in G.nodes():
        for i in range(node + 1, N):
            if p >= np.random.uniform(0, 1):
                G.add_edge(node, i)
    return G
```

```python
nx.draw(random_graph(20, 0.9))
```

### Watts-Strogatz model

```python
def watts_model(N, p):
    G = nx.Graph()
    G.add_node(0)
    G.add_node(1)
    G.add_edge(0, 1)
    for i in range(2, N):
        G.add_node(i)
        G.add_edge(i-1, i)
        G.add_edge(i-2, i)
    G.add_edge(N-1, 0)
    G.add_edge(N-1, 1)
    G.add_edge(N-2, 0)
    for edge in G.edges():
        if p >= np.random.uniform(0, 1):
            G.remove_edge(edge[0], edge[1])
            node = random.choice(list(G.nodes()))
            G.add_edge(edge[0], node)
    return G
```

```python
N = 10
plt.figure(figsize=(16,8))
for p in np.linspace(0, 1, 11):
    plt.subplot(4, 3, int(p*10) + 1)
    plt.title("p = {}".format(round(p, 1)))
    nx.draw_circular(watts_model(N, p))
```

### Barabasi-Albert model

```python
def Barabasi_model(N, steps, links):
    G = nx.Graph()
    G.add_node(0)
    for i in range(1, N):
        G.add_node(i)
        G.add_edge(i-1, i)
    G.add_edge(N-1, 0)
    for i in range(1, steps + 1):
        nodes = random.sample(list(G.nodes()), k = links)
        G.add_node(N + i)
        for j in range(links):
            G.add_edge(N + i, nodes[j])
    return G
```

```python
nx.draw(Barabasi_model(10, 5, 2))
```

```python
nx.draw(Barabasi_model(10, 5, 3))
```

```python
nx.draw(Barabasi_model(3, 10, 2))
```

### Statistic for random graph

Number of nodes

```python
len(G1.nodes())
```

Average degree

```python
np.mean([k[1] for k in list(G1.degree)])
```

Histogram of a degree

```python
degrees = [k[1] for k in list(G1.degree)]
freq = [degrees.count(degrees[i])/len(degrees) for i in range(len(degrees))]
```

```python
plt.scatter(degrees, freq)
plt.show()
```
